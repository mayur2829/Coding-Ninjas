```python
#Iris Values

import pandas as pd
columns_names=["SepalLength" , "SepalWidth" , "PetalLength" , "PetalWidth" , "Species" ] 
iris=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',names=columns_names)

Irissetosa = iris[iris['Species'] == 'Iris-setosa']
Irisversicolor = iris[iris['Species'] == 'Iris-versicolor']
Irisvirginica = iris[iris['Species'] == 'Iris-virginica']
setoamin = Irissetosa[['SepalLength','SepalWidth','PetalLength','PetalWidth','Species']].min()
setoamax = Irissetosa[['SepalLength','SepalWidth','PetalLength','PetalWidth','Species']].max()
setoaavg = Irissetosa[['SepalLength','SepalWidth','PetalLength','PetalWidth','Species']].mean()

virginicamin = Irisvirginica[['SepalLength','SepalWidth','PetalLength','PetalWidth','Species']].min()
virginicamax = Irisvirginica[['SepalLength','SepalWidth','PetalLength','PetalWidth','Species']].max()
virginicaavg = Irisvirginica[['SepalLength','SepalWidth','PetalLength','PetalWidth','Species']].mean()

versicolormin = Irisversicolor[['SepalLength','SepalWidth','PetalLength','PetalWidth','Species']].min()
versicolormax = Irisversicolor[['SepalLength','SepalWidth','PetalLength','PetalWidth','Species']].max()
versicoloravg = Irisversicolor[['SepalLength','SepalWidth','PetalLength','PetalWidth','Species']].mean()

for i in setoamin:
    print(format(i, '.2f'),end=" ")
print()
for i in setoamax:
    print(format(i, '.2f'),end=" ")
print()
for i in setoaavg:
    i = round(i,2)
    print(i,end=" ")
print("Iris-setosa")
for i in virginicamin:
    print(i,end=" ")
print()
for i in virginicamax:
    print(i,end=" ")
print()
for i in virginicaavg:
    i = round(i,2)
    print(i,end=" ")
print("Iris-versicolor")
for i in versicolormin:
    print(i,end=" ")
print()
for i in versicolormax:
    print(i,end=" ")
print()
for i in versicoloravg:
    i = round(i,2)
    print(i,end=" ")
print("Iris-versicolor")
```

```python
#Terror Frequency

import pandas as pd

import numpy as np

df_terrorism=pd.read_csv('terrorismData.csv',encoding="utf8")

year=len(set(df_terrorism['Year']))

df_terrorism=df_terrorism[df_terrorism['Country']=='India']

df_terrorism['Casualty']=df_terrorism['Killed']+df_terrorism['Wounded']

Jammu_state=df_terrorism[df_terrorism['State']=='Jammu and Kashmir']

red_state=df_terrorism[(df_terrorism['State']=='Jharkhand')|(df_terrorism['State']=='Odisha')|(df_terrorism['State']=='Andhra Pradesh')|(df_terrorism['State']=='Chhattisgarh')]

red_casualty=int(np.sum(red_state['Casualty']))

Jammu_casualty=int(np.sum(Jammu_state['Casualty']))

print(red_casualty//year,Jammu_casualty//year)
```

```python
#Terror Government

import pandas as pd
terData = pd.read_csv('terrorismData.csv',encoding="utf8")
df = terData.copy()
gov_df = df[(df.Country=='India') & (df.Year >= 2014) & (df.Month > 0)]
gov_df['YearMonth'] = gov_df.Year * 10000 + gov_df.Month*100 + gov_df.Day
gov_df              = gov_df[(gov_df.YearMonth >= 20140526)]
attacks   = gov_df.YearMonth.count()
att_counts = gov_df['Group'].value_counts()
for group in att_counts.keys():
    if group != 'Unknown':
        act_group = group
        break
print(int(attacks), act_group)
```
