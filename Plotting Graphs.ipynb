```python
#Microsoft Bubble Graph
import matplotlib.pyplot as plt
import numpy as np



employees=[61,71,79,91,93,89,90,94,99,128,118,114,124,131]
year=[2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
revenue=[39.79,44.28,51.12,60.42,58.44,62.48,69.94,73.72,77.85,86.83,93.58,85.32,89.95,110.36]

emp = np.arange(len(employees))

plt.scatter(year,revenue, s=employees, c = emp)
plt.title("Bubble Chart")
plt.show()

for i in range(1,len(employees)-1):
    if revenue[i] > revenue[i-1] and revenue[i] > revenue[i+1]:
        print(year[i],revenue[i],employees[i])
```

# Pie chart

```python

labels = ["A","B","C","D"]
sizes = [3,4,6,2]
colors = ["blue","red","orange","grey"]
explode = [0.1,0.2,0,0]
plt.title("Split of classes")

plt.pie(sizes, colors=colors, explode = explode, labels= labels, autopct="%.2f%%", counterclock = False, startangle=90)
plt.axis("equal")

plt.show()
```

```python
#Laptop Sales

import matplotlib.pyplot as plt
import math

company=['HP','Dell','Lenovo','Asus','Apple','Acer']
sold=[20000,43000,15000,17000,22000,13000]
colors = ["blue","red","orange","grey","pink","green"]

sum = sum(sold)

values=[]
for s in sold:
    values.append((s/sum)*100)


plt.pie(sold, colors=colors, labels= company, autopct="%.1f%%")

plt.show()

for i in range(0,len(sold)):
    print(company[i], "{0:.1f}".format( values[i]))
```

```python
#Plot Jobs

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('amazon_jobs_dataset.csv')

df["Year"] = df.Posting_date.str.split(",",n = 1,expand=True)[1]
k = df["Year"].value_counts()[::-1]

years = k.keys()
n_yrs = np.array(years)
jobs = k.values
n_jobs =np.array(jobs)
k_ind = n_yrs.argsort()

n_jobs =n_jobs[k_ind]
n_yrs = n_yrs[k_ind]

plt.plot(n_yrs,n_jobs)
plt.show()
for i in range(len(years)):
    print(int(n_yrs[i]),int(n_jobs[i]))
```

```python
#Plot Month_Job

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('amazon_jobs_dataset.csv')
temp = df.Posting_date.str.split(",",n = 1,expand=True)[0]
month = temp.str.split(" ",n = 1,expand = True)[0]
freq = month.value_counts()
x = freq.index
y = freq
plt.bar(x,y)
plt.xticks(rotation = 40)
plt.show()
for i in range(len(x)):
    print(x[i],y[i])
```

```python
#Plot city_job


import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('amazon_jobs_dataset.csv')
df["Country"] = df.location.str.split(",",expand= True)[0]
df = df[df["Country"] == "IN"]
df["City"] = df.location.str.split(",",expand=True)[2]
freq = df["City"].value_counts()
freq = freq.sort_values(ascending=False)
city = freq.index
jobs = freq
total = freq.sum()

plt.pie(jobs,labels=city,autopct="%.2f%%")
plt.axis("equal")
plt.show()

for i in range(len(freq)):
    perc = (jobs[i]/total)*100
    print(city[i].strip(),format(perc,".2f"))
```
