```python
#Job Language

import csv
path='amazon_jobs_dataset.csv'
country_basic_qualifications=[]
with open(path,'r',encoding='utf-8') as csvFile:
    reader=csv.DictReader(csvFile)
    for row in reader:
        country_basic_qualifications.append([row['location'],row['BASIC QUALIFICATIONS']])
## Shortlist the no of job opening for India
def getCountryJobOpening(arr,country_code):
    arrlist=[]
    for i in arr:
        if i[0].split(',')[0]==country_code:
            arrlist.append(i[1])
    return arrlist
india_list=getCountryJobOpening(country_basic_qualifications,'IN')
## Now find the job for bachelor degree holder
def getJobBachelorList(arr):
    arrlist=[]
    for i in arr:
        if 'Bachelor' in i or 'BS' in i or 'BA' in i:
            arrlist.append(i)
    return arrlist
india_bachelor_list=getJobBachelorList(india_list)
## Get specificed Language
def getTheDeveloper(arr,language):
    arrlist=[]
    for i in arr:
        if language in i:
            arrlist.append(i)
    return arrlist
java=[len(getTheDeveloper(india_bachelor_list,'Java')),'Java']
Cplus=[len(getTheDeveloper(india_bachelor_list,'C++')),'C++']
py=[len(getTheDeveloper(india_bachelor_list,'Python')),'Python']
maximum=max(java,Cplus,py)
print(maximum[1],maximum[0])
```

```python
#Job Java

import csv
path='amazon_jobs_dataset.csv'
country_basic_qualifications=[]
with open(path,'r',encoding='utf-8') as csvFile:
    reader=csv.DictReader(csvFile)
    for row in reader:
        country_basic_qualifications.append([row['location'],row['BASIC QUALIFICATIONS']])
## Get Java Developer from the different country
def getTheJavaDeveloper(arr):
    arrlist=[]
    for i in arr:
        if 'Java' in i[1]:
            arrlist.append(i[0].split(',')[0])
    return arrlist
java_developer=getTheJavaDeveloper(country_basic_qualifications)
def createDictionary(arr):
    dictionary={i:0 for i in set(arr)}
    for i in arr:
        dictionary[i]=dictionary.get(i)+1
    return dictionary
def convertDictToList(dictionary):
    arrlist=[]
    for i in dictionary:
        arrlist.append([dictionary.get(i),i])
    arrlist.sort(reverse=True)
    return arrlist
java_dict=createDictionary(java_developer)
java_list=convertDictToList(java_dict)
print(java_list[0][1],java_list[0][0])
```
